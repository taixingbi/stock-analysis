```python
def get_return_tot(df):
    threshold = -0.01
    return_tot = ""
    indexes = df.index.to_list()
    for index in indexes:
        if df.loc[index]['pct_change'] < threshold : 
            end_price = df.loc[index]['open_price']
            start_price = df.loc[0]['open_price']
            return_tot_float = 1.0*100*(end_price-start_price)/start_price
            if return_tot_float < 3: return False, return_tot
            return_tot = round(return_tot_float, 2)
            break

    return True, str(return_tot) + "%"

def validfilter(name = 'NRDS'):
    d = rs.stocks.get_stock_historicals(name, interval="5minute", span="day") # "5minute","10minute","hour","day",or "week"
    df = None
    df = pd.DataFrame(data=d)

    df['open_price']= round(df['open_price'].astype(float),2)
    df['pct_change'] = df['open_price'].pct_change()
    df = df[ df['volume'] > 0  ]
    if len(df) ==0 : 
        print("volume is zero")
        return
    
    df = df.reset_index()
    df.index = df.index * 5
    for m in range(15):
        if m%5 == 0:
            if df['pct_change'][m] < 0 : return
            
    valid, return_tot = get_return_tot(df)
    if not valid: return
    df['open_price'].plot( title= name + " " + return_tot )
    plt.show()
    
# tickers = ['NRDS']
for name in tickers[:]:
    try:
        validfilter(name)
    except:
        print(name)
        continue
```
